```python
#Parse HDF5 File

import h5py
import numpy as np

def h5reader(h5file): 
	# input: PATH to the download h5file: highly recommend to use absolute PATH for h5file
	f = h5py.File(h5file, 'r')
	timestep = f['timestep'].value
    
	bodies = {}
	body_field_names= f['bodies'].items()
	print("******************************************************")
	print("These are the bodies field names and field values")
	for key, val in body_field_names:
	    print(key, '--->', val.value)
	    bodies[key] = val.value
	    
	contacts = {}
	contact_field_names= f['contacts'].items()
	print("******************************************************")
	print("These are the contacts field names and field values")
	for key, val in contact_field_names:
	    print(key, '--->', val.value)
	    contacts[key] = val.value
	
	constraints = {}
	constraint_field_names= f['constraints'].items()
	print("******************************************************")
	print( "These are the constraint field names and field values")
	for key, val in constraint_field_names:
	    print(key, '--->', val.value)
	    constraints[key] = val.value
        
```
